Ans- Tuple items are ordered, unchangeable, and allow duplicate values. Tuple items are indexed, the first item has index [0] , the second item has index [1] etc.We can also define tuples as lists that we cannot change. Therefore, we can say yes tuples are immutable.

Ans- Tuple has 2 functions: count() and index(). 
>Count() gives total occurences of a specific element
Index() gives location of first occurence of an element
Tuples has only two methods as it is immutable so methods like adding, deleting, updating etc are not available for tuples

```python
#example for tuple method
Tuple=[1,2,3,4,5,6,7,8,'abc',5]
print("the count of 5 is:",Tuple.count(5))
print("the index of abc is:",Tuple.index('abc'))
```

Ans- The update() method updates the original set by adding items from all the specified sets, with no duplicates whereas set. union() creates a new set and returns a copy.

```python
set_2={2,5,1,84,32}
set_3={34,2,56,7,1}
set_2.update(set_3)#update the original set by adding items from specified set
print(set_2)
set_4=set_2.union(set_3)
print(set_4)
```

Python Dictionary setdefault() returns the value of a key (if the key is in dictionary). Else, it inserts a key with the default value to the dictionary.
>Syntax: dict.setdefault(key, default_value)


```python
dict1 = {'language' : 'Python',
         'course': 'Data Science Masters',
        }
topics = dict1.setdefault('topics', ['Python', 'Machine Learning', 'Deep Learning'])
dict1
```
