 Answer-'while' loop in Python is used to execute a set of statements repeatedly as long as a
condition is true. The loop continues to execute untill the condition becomes false. The
basic syntax of a 'while' loop is as follow:
           
              while condition:
                   #statement
              else:
                   #statement
            
where 'condition' is an expression that returns either 'True' or'False'. If the condition 
is true, the loop will execute. If the condition is false, the loop will terminate and the 
control will be passed to the next statement after the loop. It is important to note that if 
the condition in 'while' loop is never statisfied, the loop will run indefinitely, ausing an 
infinite loop. To avoid this we must make sure that the condition in 'while' loop eventually becomes false. Example of a while loop is given below:

```python
#while loop to print 1 to 10
i=1
while i<=10:
        print(i)
        i+=1
```

```python
#right angle triangle pattern using nested while loop
print("right angle triangle pattern ")  
i=1
while i<6:
    j=1
    while j in range(j,i+1):
        print ('*',end="")
        j=j+1
    print("\r")
    i=i+1
```

```python
#reversed triangle pattern using nested while loop
print("reversed triangle pattern ")  
rows = 5
i = rows
while i >= 1:
    j = 1
    while j <= i:
        print("*", end="")
        j = j + 1
    print("")
    i = i - 1
```

```python
#pyramid pattern using nested while loop
print("pyramid pattern ")  
rows = 5
for i in range(1, rows + 1):
    for j in range(rows - i):
        print(end=" ")
    for j in range(2 * i - 1):
        print("*", end="")
    print("")
```
