Ans- <b>Line Plot-</b> This Plot Helps Us To See The Relationship Between X-Axis, Y-Axis And It Also Takes Some Parameters Such As Hue, Size, Color, Etc.  </br>

<b>Countplot-</b>  ShowsThe Counts Of Observations In Each Categorical Bin Using Bars.This Method Is Accepting The Parameters X, Y  Which Take The Name Of A Variable In Data, Hue It Is An Optional Parameter It Helps To Take Column Name For Color Encoding. Data Is Also An Optional Parameter That Takes DataFrame, Array, Or List Of Arrays, Dataset For Plotting.

<b>Bar Chart-</b> A Bar Chart Is A Way Of Comparing A Set Of Categorical Data. It Is Better To Convert Continuous Data To Bins Before Plotting. The Bar Chart Displays Data Using Several Bars, Each Representing A Particular Category. This Method Is Accepting The Parameters X, Y  Which Take The Name Of A Variable In Data, Hue It Is An Optional Parameter It Helps To Take Column Name For Color Encoding.

<b>Pairplot-</b> Pair plot is used to understand the best set of features to explain a relationship between two variables or to form the most separated clusters. It also helps to form some simple classification models by drawing some simple lines or make linear separation in our data-set.

<b>Scatter Plot-</b> Scatter plots are the graphs that present the relationship between two variables in a data-set. It represents data points on a two-dimensional plane or on a Cartesian system. The independent variable or attribute is plotted on the X-axis, while the dependent variable is plotted on the Y-axis.

<b>Histogram-</b> A histogram is a bar graph-like representation of data that buckets a range of classes into columns along the horizontal x-axis. The vertical y-axis represents the number count or percentage of occurrences in the data for each column. Columns can be used to visualize patterns of data distributions.

```python
#importing the seaborn library
import seaborn as sns
df = sns.load_dataset('fmri')#loading the fmri dataset
df.head(4)
```

```python
df.isnull().sum()
```

```python
sns.lineplot(data=df, x="timepoint",y="signal",hue="event",style="region")
```

```python
#importing the seaborn library
df_2 = sns.load_dataset('diamonds') #loading the titanic dataset
df_2.head(4)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,6))
sns.histplot(data=df_2,x="price",hue="cut",kde=True)
plt.title("distribution of price for different type of diamonds cut")
plt.show()
```
